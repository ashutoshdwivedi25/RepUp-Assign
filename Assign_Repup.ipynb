Splitting Training and testing data:

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.25, random_state=53)
```

```python
X_train.head()
```

```python
X_test.head()
```

Using CountVectorizer and tfidf Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
count_vectorizer = CountVectorizer(stop_words='english')
count_train = count_vectorizer.fit_transform(X_train['News']) 
count_test = count_vectorizer.transform(X_test['News'])
```

```python
print(count_train.A.shape)
print(X_train.shape)
print(len(count_vectorizer.get_feature_names()))
```

```python
count_df = pd.DataFrame(count_train.A, columns=count_vectorizer.get_feature_names())
```

```python
count_df.head()
```

# using multinomial naive bayes and countvectorizer


```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB() 
```

```python
from sklearn import metrics
```

```python
clf.fit(count_train, y_train)
pred = clf.predict(count_test)
score = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % score)
```

# Using Passive Aggressive Classifier and Countvectorizer

```python
from sklearn.linear_model import PassiveAggressiveClassifier
```

```python
linear_clf = PassiveAggressiveClassifier(n_iter=50)
```

```python
linear_clf.fit(count_train, y_train)
pred = linear_clf.predict(count_test)
score = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % score)
```

Using Tf-IDF

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7) 
```

```python
tfidf_train = tfidf_vectorizer.fit_transform(X_train['News']) 
```

```python
tfidf_test = tfidf_vectorizer.transform(X_test['News'])
```

```python
tfidf_df = pd.DataFrame(tfidf_train.A, columns=tfidf_vectorizer.get_feature_names())
```

```python
tfidf_df.head()
```

# using multinomial naive bayes and tfidf

```python
clf = MultinomialNB() 
```

```python
clf.fit(tfidf_train, y_train)
pred = clf.predict(tfidf_test)
score = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % score)
```

# Using Passive Aggressive Classifier and tfidf

```python
linear_clf = PassiveAggressiveClassifier(n_iter=100)
```

```python
linear_clf.fit(tfidf_train, y_train)
pred = linear_clf.predict(tfidf_test)
score = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % score)
```
